```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Завдання 1 - 4
df_xls = pd.read_excel('./Lab3_Datasetі/En_In.xls', skiprows=17, skipfooter=38)
df_xls = df_xls.drop(df_xls.columns[[0, 1]], axis=1)
df_xls.index = range(1, len(df_xls) + 1)
df_xls.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

df_xls['Energy Supply'] *= 1000000
df_xls[df_xls.columns[1::]] = df_xls[df_xls.columns[1::]].replace(r'\D+', np.nan, regex=True)
pd.set_option('display.float_format', '{:.3f}'.format)

df_xls['Country'] = df_xls['Country'].str.replace(r"\d+", "", regex=True)
df_xls['Country'] = df_xls['Country'].str.replace(r"\(.*?\)", "", regex=True)
df_xls['Country'] = df_xls['Country'].str.strip()

new_names = {
    "Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
}

df_xls['Country'].replace(new_names, inplace=True)

df_xls.loc[df_xls['Country'].isin(['American Samoa', 'South Korea', 'Bolivia'])]
```

```python
# Завдання 5

df_csv = pd.read_csv('./Lab3_Datasetі/gpd.csv', skiprows=4)

countries = {
    "Korea, Rep.": "South Korea",
    "Iran, Islamic Rep.": "Iran",
    "Hong Kong SAR, China": "Hong Kong"
}

df_csv['Country Name'] = df_csv['Country Name'].replace(countries)
df_csv = df_csv.rename(columns={'Country Name': 'Country'})

df_csv.head(1)
```

```python
# Завдання 6 - 7

df_xlsx = pd.read_excel('./Lab3_Datasetі/scimagojr.xlsx')

df = pd.merge(df_xls, df_csv, on='Country', how='inner')
df = pd.merge(df, df_xlsx, on='Country', how='inner')

columns = ['Rank', 'Documents', 'Citable documents','Citations', 'Self-citations','Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
df = df[df['Rank'] <= 15]
df = df.set_index('Country')
df = df[columns]
df = df.sort_values(by='Rank')

print(df.shape)
df.head(3)
```

```python
# Завдання 8

def task_eight(df):
    col_names = ['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    return df[col_names].mean(axis=1).sort_values(ascending=False)

task_eight(df)
```

```python
# Завдання 9

def task_nine(df):
    country_name = task_eight(df).index[4]
    diff = df.at[country_name, '2015'] - df.at[country_name, '2006']
    return (country_name, diff)

task_nine(df)
```

```python
# Завдання 10

def task_ten(df):
    return (df['% Renewable'].idxmax(), df['% Renewable'].max())

task_ten(df)
```

```python
# Завдання 11

def task_eleven(df):
    df['Population Estimate'] = df['Energy Supply'] / df['Energy Supply per Capita']
    sorted = df.sort_values(by='Population Estimate', ascending=False)
    return (sorted.index[5], sorted['Population Estimate'].values[5])

task_eleven(df)
```

```python
# Завдання 12

def task_twelve(df):
    df['Citable docs per Person'] = df['Citable documents'] / df['Population Estimate']
    return df['Citable docs per Person'].corr(df['Energy Supply per Capita'])

task_twelve(df)
```

```python
# Завдання 13

def task_thirteen(df):
    median_renewable = df['% Renewable'].median()
    return df['% Renewable'] >= median_renewable

task_thirteen(df)
```

```python
# Завдання 14

def task_fourteen(df):
    continents = {
        'China':'Asia',
        'United States':'North America',
        'Japan':'Asia',
        'United Kingdom':'Europe',
        'Russian Federation':'Europe',
        'Canada':'North America',
        'Germany':'Europe',
        'India':'Asia',
        'France':'Europe',
        'South Korea':'Asia',
        'Italy':'Europe',
        'Spain':'Europe',
        'Iran':'Asia',
        'Australia':'Australia',
        'Brazil':'South America'
    }

    df['Continent'] = df.index.to_series().map(continents)
    return df.groupby('Continent')['Population Estimate'].agg(['size', 'sum', 'mean', 'std'])

task_fourteen(df)
```

```python
# Завдання 15

df['GDP 2015 Normalized'] = (df['2015'] / df['2015'].max()) * 100
plt.figure(figsize=(20, 6))

for continent in df['Continent'].unique():
    continent_data = df[df['Continent'] == continent]
    
    plt.scatter(
        continent_data['Rank'],
        continent_data['% Renewable'],
        s=continent_data['GDP 2015 Normalized'] * 70,
        label=continent
    )

    for name, row in continent_data.iterrows():
        plt.text(row['Rank'], row['% Renewable'], name, fontsize=14, ha='center', va='center')

plt.xlabel('Rank')
plt.ylabel('% Renewable')
plt.title('Bubble chart')
plt.legend()
plt.show()
```
